```python
import numpy as np 
import matplotlib.pyplot as plt 
import os 
plt.rcParams['font.family']="serif"
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True

plt.figure(figsize=(7,3))

lwl =  0.707106
lgl = 1.5731322 - lwl 
lgx = 2.5731322 - lwl - lgl 
lxw = 3.0731322 - lwl - lgl - lgx 

data = np.loadtxt('e-psi-lif-hole.dat') 

xk_len = data[0] 
ex = np.zeros([31,3])
pt_size = np.zeros([31,3])
ex[:,0] = data[1]
ex[:,1] = data[2]
ex[:,2] = data[3]
pt_size[:,0] = data[4]
pt_size[:,1] = data[5]
pt_size[:,2] = data[6]


plt.plot(xk_len, ex,c='k',alpha=0.8)


for i in range(3):
   #plt.scatter(xk_len[::3], ex[::3,i], s = psi_x[::3,i]*1e8,c='tab:blue',edgecolor='k')
   plt.scatter(xk_len, ex[:,i], s = pt_size[:,i],c='tab:red',edgecolor='k',alpha=0.7)
plt.xlim([xk_len[0],xk_len[-1]])
plt.xticks([0,lwl,lwl+lgl,lwl+lgl+lgx, lwl+lgl+lgx+lxw],['W','L','G','X','W'],fontsize=24)
plt.axvline(lwl,zorder=-1,c='k',lw=1,alpha=0.7)
plt.axvline(lwl+lgl,zorder=-1,c='k',lw=1,alpha=0.7)
plt.axvline(lwl+lgl+lgx,zorder=-1,c='k',lw=1,alpha=0.7)

plt.yticks([-3,-2,-1,0],fontsize=24)
plt.minorticks_on()
plt.ylabel('Energy (eV)', fontsize=25)

plt.ylim([-3.5,0.4])
plt.tight_layout()
plt.savefig('fig-3a-1.pdf',dpi=500)

```

```python
import numpy as np 
import matplotlib.pyplot as plt
import os  
plt.rcParams['font.family']="serif"
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True


plt.figure(figsize=(7,5))
Znph = np.loadtxt('Znph-lif-hole.dat')
plt.bar(np.arange(len(Znph)),Znph)
plt.xlim([0,80])
plt.ylim([0,0.15])
plt.xticks([0,10,20,30,40,50,60,70],fontsize=24)
plt.minorticks_on()
plt.yticks([0,0.04,0.08,0.12],fontsize=24)
plt.xlabel('N phonons', fontsize=25)
plt.ylabel(r'$Z_N$', fontsize=30)
plt.tight_layout()
plt.savefig('fig-3a-2.pdf',dpi=500)

N_ph = np.sum(np.arange(len(Znph))*Znph) / np.sum(Znph)
print('Total-phonon number = ', N_ph)

```
