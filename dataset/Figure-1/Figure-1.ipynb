```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.optimize as opt 
import os 

plt.rcParams['font.family']="sans-serif"
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True



fig, ax = plt.subplots(figsize=(7,5))


data = np.loadtxt('lif-gt-band.dat')
tau = data[0]
trG = data[1]
dtrG = data[2]

ax.plot(tau[:-1],trG[:-1], '-',c='tab:orange',lw=3,label='Band')
ax.fill_between(tau, trG - dtrG, trG + dtrG,color='tab:orange', alpha=0.4)



data = np.loadtxt('lif-gt-matrix.dat')
tau = data[0]
trG = data[1] 
dtrG = data[2] 
plt.plot(tau[:-1], trG[:-1],'--',lw=3,c='tab:blue',label='Matrix product')
ax.fill_between(tau, trG - dtrG, trG + dtrG,color='tab:blue', alpha=0.4)

plt.xlabel(r'$\tau$ (eV$^{-1}$)',fontsize=25)
plt.ylabel(r'$G(\tau)$',fontsize=25)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.xlim([0,12])
plt.ylim([-0.5,1.4])
plt.axhline(0,lw=1,c='k',zorder=-1)
plt.legend(fontsize=20,frameon=False)
plt.tight_layout()
plt.savefig('error-Gtau-a=1.0.pdf',dpi=500)
```
