```python
import numpy as np
import matplotlib.pyplot as plt 
import os
from matplotlib import rcParams
plt.rcParams['font.family']="serif"
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
lgz = 0.3977256
lgx = 1.1048324 - lgz



x = np.loadtxt('Ek-anatase.dat') 


plt.figure(figsize=(7.5,5))
plt.errorbar(x[0],x[1],yerr=x[2],lw=2.5, capsize=4,c='tab:orange',label='DMC')

x = np.loadtxt('tio2-gw-elec.bands')
nbnd = 10
nk = int(len(x)/nbnd)
print(len(x), nk*nbnd)
xk = x[:,0].reshape([nbnd,nk])
ek = x[:,-1].reshape([nbnd,nk])
CBM = np.min(ek) 

for ib in range(nbnd):
   plt.plot(xk[ib],ek[ib]-CBM,c='tab:blue',lw=3)
plt.plot(xk[ib],ek[ib]-CBM,c='tab:blue',lw=3,label='DFT')


plt.xlim([0,lgx+lgz])
plt.ylim([-0.3,0.5])
plt.xticks([0,lgz,lgz+lgx],['Z','G','X'],fontsize=30)
#plt.grid()
plt.minorticks_on()
plt.tick_params( which='minor', direction='in')
plt.tick_params( which='major', direction='in')
plt.axvline(lgz,c='k',lw=1,alpha=0.6)
plt.axhline(0,c='k',lw=1,alpha=0.6)
plt.legend(fontsize=20,frameon=False)
plt.yticks([-0.2,0,0.2,0.4],fontsize=25)

plt.ylabel(r'Energy (eV)',fontsize=30)
plt.tight_layout()


plt.savefig('fig-4d.pdf', dpi= 500)
```
